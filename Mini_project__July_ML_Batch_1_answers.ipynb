## Which are the movies with the third lowest and third highest budget?  

```python
Third_lowest = movies.sort_values(['budget']).iloc[2]['original_title']
Third_highest = movies.sort_values(['budget']).iloc[-3]['original_title']
print('Third lowest : ', Third_lowest)
print('Third highest :' ,Third_highest) 
```

##  What is the average number of words in movie titles between the year 2000-2005?

```python
bw_00_05 = movies[movies['release_year'].isin([2000, 2001, 2002, 2003, 2004, 2005])]
```

```python
nwords = 0
for i in range(bw_00_05.shape[0]):
    words_list= bw_00_05['original_title'].values[i].split(' ')
    nwords = nwords + len(words_list)

average=nwords/bw_00_05.shape[0]
print('Average number of words in movie titles between the year 2000-2005 is :  ',average)
```

##   What is the most common Genre for Vin Diesel & Emma Watson movies?

```python
vindiesel = movies[movies['cast'].str.contains('Vin Diesel')]
emmawatson = movies[movies['cast'].str.contains('Emma Watson')]
```

```python
genres_and_count = {}
for i in range(vindiesel.shape[0]):
    genres = vindiesel['genres'].values[i].split('|')
    for genre in genres:
        try:
            count = genres_and_count[genre]
            genres_and_count[genre] = count + 1
        except:
            genres_and_count[genre] = 1
genres_and_count
```

##  Which are the movies with most and least earned revenue?

## What is the average runtime of movies in the year 2006?

```python
df_2006 = movies[movies['release_year'] == 2006]
total_runtime = 0
for i in range(df_2006.shape[0]):
    total_runtime += int(df_2006['runtime'].values[i])
    
average= total_runtime/df_2006.shape[0]
print('The average runtime of movies in year 2006 is :',average)
```
